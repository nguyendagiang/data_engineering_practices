Data File - Excel

```python
# input variables
excel_file_path = "auditor_and_liquidator.xlsx"
auditor_sheetname = "REG_AUDITOR_202511"
liquidator_sheetname = "LIQUIDATOR_202510"

# output variables
# path = "data_lake"#reserved for data lake
```

Read Auditor data

```python
# approach 1: load data into data frame then use astype to convert data type of specific columns 
import pandas as pd
import openpyxl 
import datetime
import os

dtype_dict={
    'REG_AUD_ACN':'string',
    'REG_AUD_START_DT':'datetime64[ns]',
    'REG_AUD_SUSP_DT':'datetime64[ns]',
    'REG_AUD_ADD_PCODE':'string',
    'REG_AUD_NUM':'string'
}

# 1. Read data from "REG_AUDITOR_202511" spreadsheet using Pandas
df = pd.read_excel(excel_file_path ,sheet_name=auditor_sheetname,header=2 )

# 2. Apply data type conversion and export to CSV
# changing specific columns using a dictionary
df=df.astype(dtype_dict)

# Check the data types of the DataFrame after converting)
print(df.dtypes)


# 3. Export to path 'destination/excel'
# Save the DataFrame to a new Excel file

export_date=datetime.datetime.now()
print(export_date)
export_date=export_date.strftime("%Y%m%d")
folder_path=f'destination/excel/{export_date}'
os.makedirs(folder_path,exist_ok=True)
print(f'Exporting {excel_file_path}...')
export_file_name=export_date + '_' + auditor_sheetname + '.xlsx' 
df.to_excel(f'{folder_path}/{export_file_name}', index=False) # index=False prevents writing row indices to Excel
print('Completed!')
 
```

```python
# approach 2: predefine data type dictionary, convert data type on reading 

# 1. Read data from "LIQUIDATOR_202510" spreadsheet using Pandas
#    + Provide a schema when reading data file
# liquidator_schema = {
#     "REGISTER_NAME": str,
#     "LIQ_NUM": str,
#     "OFF_LIQ_NUM": str, 
#     "LIQ_NAME": str,
#     "LIQ_START_DT": 'datetime64[ns]',
#     "OFF_LIQ_START_DT": 'datetime64[ns]',
#     "LIQ_STATUS": str,
#     "LIQ_SUSP_DT": 'datetime64[ns]',
#     "LIQ_ADD_LOCAL": str,
#     "LIQ_ADD_STATE": str,
#     "LIQ_ADD_PCODE": str,
#     "LIQ_ADD_COUNTRY": str,
#     "LIQ_FIRM": str
# }
# 2. Export to CSV as path 'destination/excel'


import pandas as pd
import openpyxl 
import datetime
import os

# define data type dictionary for specific columns
liquidator_schema = {
    "REGISTER_NAME": str,
    "LIQ_NUM": str,
    "OFF_LIQ_NUM": str, 
    "LIQ_NAME": str,
    "LIQ_START_DT": 'datetime64[ns]',
    "OFF_LIQ_START_DT": 'datetime64[ns]',
    "LIQ_STATUS": str,
    "LIQ_SUSP_DT": 'datetime64[ns]',
    "LIQ_ADD_LOCAL": str,
    "LIQ_ADD_STATE": str,
    "LIQ_ADD_PCODE": str,
    "LIQ_ADD_COUNTRY": str,
    "LIQ_FIRM": str
}

# 1. Read data from "REG_AUDITOR_202511" spreadsheet using Pandas
# 2. Apply data type conversion and export to CSV

# Read the Excel file, applying the specified data types
df = pd.read_excel(excel_file_path ,sheet_name=liquidator_sheetname,header=2,dtype=liquidator_schema )

# Check the data types of the DataFrame columns
print(df.dtypes)


# 3. Export to path 'destination/excel'
# Save the DataFrame to a new Excel file

export_date=datetime.datetime.now()
print(export_date)
export_date=export_date.strftime("%Y%m%d")
folder_path=f'destination/excel/{export_date}'
os.makedirs(folder_path,exist_ok=True)
print(f'Exporting {excel_file_path}...')
export_file_name=export_date + '_' + liquidator_sheetname + '.xlsx' 
df.to_excel(f'{folder_path}/{export_file_name}', index=False) # index=False prevents writing row indices to Excel
```
